```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from bs4 import BeautifulSoup as bs
import time

# inicializar o driver do Selenium
driver = webdriver.Chrome()

url = 'https://www.econodata.com.br/maiores-empresas/todo-brasil'

headers = {
    'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Safari/537.36 Edg/105.0.0.0"
}

# selecionar a opção de 500 empresas
driver.get(url)
select = Select(driver.find_element(By.ID, "opcoes-tamanho"))
select.select_by_value("500")

# lista para armazenar as empresas
empresas = []

# percorrer todas as páginas
max_pag = 50
page = 1
while page <= max_pag:
    # aguardar a tabela de empresas ser carregada
    table_element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//tbody[@role='rowgroup']"))
    )
    
    # encontrar a tabela com as empresas
    soup = bs(driver.page_source, 'html.parser')
    table = soup.find('tbody', {'role': 'rowgroup'})

    # percorrer todas as linhas da tabela
    for row in table.find_all('tr', {'role': 'row'}):
        # extrair os dados de cada empresa
        cols = row.find_all('td', {'role': 'cell'})
        ranking = cols[0].find('div', {'class': 'ranking-number'}).text.strip()
        nome = cols[1].find('a').text
        cnpj = cols[1].find('div', {'class': 'table-font-small'}).text.strip()[0:18]
        endereco = cols[2].find('div', {'class': 'table-cell-endereco'}).text.strip()
        cidade_estado = endereco.split('\n')[-1].strip()
        cidade = cidade_estado.split(', ')[0]
        estado = cidade_estado.split(', ')[1]
        setor = cols[3].find_all('div')[2].text.strip()
        faturamento = cols[3].find_all('div')[-1].text.strip().replace(',', '').replace('.', '')

        # adicionar a empresa na lista
        empresas.append({'ranking': ranking, 'nome': nome, 'cnpj': cnpj, 'cidade': cidade, 'estado': estado, 'setor': setor, 'faturamento': faturamento})

        page += 1

# print(empresas)
```

```python
import pandas as pd

df = pd.DataFrame(empresas)
df = df.set_index('ranking')
df.head(20)
```
