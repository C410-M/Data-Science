```python
musics_info = []

for i, item in enumerate(songs):
    track = item['track']
    added_at = item['added_at']
    try:
      artist = sp.artist(track['artists'][0]['id'])
    except AttributeError:
       print('Erro ao obter dados do artista')
    musics_info.append((track['artists'][0]['name'], track['name'], track['album']['name'], added_at, track['id'], artist['genres'], artist['popularity']))

#print(artists)

pd.set_option('display.max_rows', 50)
musics = pd.DataFrame(musics_info, columns=('Artists', 'Music', 'Album', 'Added_At', 'Id', 'Genres', 'Popularity'))
musics
```
