```python
playlist_id = '37i9dQZF1DWY3ikieEzZoK'
offset = 0
playlist = sp.playlist_tracks(playlist_id = playlist_id, offset = offset)
songs = playlist['items']
```

```python
df = pd.DataFrame(songs)
df
```

```python
musics_info = []

for i, item in enumerate(songs):
    track = item['track']
    added_at = item['added_at']
    try:
      artist = sp.artist(track['artists'][0]['id'])
    except AttributeError:
       print('Erro ao obter dados do artista')
    musics_info.append((track['artists'][0]['name'], track['name'], track['album']['name'], added_at, track['id'], artist['genres'], artist['popularity']))

#print(artists)

pd.set_option('display.max_rows', 50)
musics = pd.DataFrame(musics_info, columns=('Artists', 'Music', 'Album', 'Added_At', 'Id', 'Genres', 'Popularity'))
musics
```

```python
af = sp.audio_features(musics['Id'])[0:100]

af = pd.DataFrame(af, columns = ['energy', 'liveness',
'tempo', 'speechiness',
'acousticness', 'instrumentalness',
'time_signature', 'danceability',
'key', 'duration_ms', 'loudness',
'valence', 'mode', 'type', 'uri'])

af
```

```python
songs_features = pd.concat([musics, af]) # Concatenando os dois Datasets
songs_features.fillna(0, inplace=True)
songs_features.set_index(['Artists'], inplace = True) # Passando a variável 'SEQN' como Index ou Chave Única
songs_features
```
